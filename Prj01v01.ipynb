```python
#create an empty list
r = list()
```

First we try to get all prime numbers in 1000~1000000. The simple method is to tests whether $ n $ is a multiple of any integer between 2 and $ \sqrt{ n } $. Since we know that a prime number should be an odd number(except 2), to save time, we just the odd numbers in 100~1000000. So we set the step length to 2.

```python
#find prime numbers between 1000 and 1000000
for a in range(1001, 1000000, 2):
    for j in range(3, int(a**0.5)+1, 2):
        if a % j == 0:
            break
    else:
        r.append(a)
```

```python
print(r)
```

```python
#get the number of these primes
n=len(r)
```

```python
print(n)
```

```python
count = 0
for i in range(0, n-2):
    if r[i+1]-r[i] == 2:
        print(r[i], r[i+1])
        count +=1
```

```python
print(count)
```
